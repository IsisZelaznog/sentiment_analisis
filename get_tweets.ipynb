```python
#Introduce tu bearer token de acceso a la API de Twitter, esto lo hicimos en el modulo 3
bearer_token = 'XXXX'
```

```python
client = tweepy.Client(bearer_token=bearer_token,wait_on_rate_limit=True)
```

```python
tweets=[]
class listener(tweepy.StreamingClient):
    def on_connect(self):
        self.counter = 0
        self.limit = 1000
    def on_tweet(self, tweet):
        tweets_stream=[]
        print(tweet.text)        
        tweets_stream.append(tweet.author_id)
        tweets_stream.append(tweet.created_at)
        tweets_stream.append(tweet.text)
        tweets_stream.append(self.counter)
        tweets.append(tweets_stream)       
        self.counter += 1
        if self.counter < self.limit:
            return True
        else:
          self.disconnect()
    def on_error(self, tweet):
        print(tweet)
```

```python
#Podeis ponerle las etiquetas que querais para filtrar
search_terms=["Ukraine War"]
listen = listener(bearer_token,wait_on_rate_limit=True)


printer_rules = listen.get_rules()
if printer_rules.data is not None:
  listen.delete_rules([str(i.id) for i in listen.get_rules().data])

#Declaramos nuevas reglas para el filtrado
for terms in search_terms:
  listen.add_rules(tweepy.StreamRule(f"{terms} -is:retweet lang:en -is:reply"))
listen.filter(tweet_fields=["author_id","created_at","text"])
```

```python
#Objeto tweets a dataframe
columns=["User_id","Created_at","Text","Count"]
df= pd.DataFrame(tweets, columns=columns)
```
